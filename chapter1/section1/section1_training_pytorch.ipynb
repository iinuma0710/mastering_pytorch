````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
"""
ネットワークの構築
 - 畳み込み層
 - ドロップアウト層
 - 線形 (全結合) 層
"""

class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.cn1 = nn.Conv2d(1, 16, 3, 1)
        self.cn2 = nn.Conv2d(16, 32, 3, 1)
        self.dp1 = nn.Dropout2d(0.10)
        self.dp2 = nn.Dropout2d(0.25)
        self.fc1 = nn.Linear(4608, 64) # 4608 = 12 x 12 x 32
        self.fc2 = nn.Linear(64, 10)
        
    def forward(self, x):
        x = self.cn1(x)
        x = F.relu(x)
        x = self.cn2(x)
        x = F.relu(x)
        x = F.max_pool2d(x, 2)
        x = self.dp1(x)
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dp2(x)
        x = self.fc2(x)
        op = F.log_softmax(x, dim=1)
        return op
````

```__init___``` 関数ではモデルのコアとなるアーキテクチャ，すなわち各層のニューロンの数を定義しています．```forward``` 関数では，名前の通りネットワークの順伝播を定義しています．したがって，各層のすべての活性化関数と，各層の後に使用されるドロップアウトやプーリングも含まれています．この関数は最終層の出力を返します．これをモデルの予測 (prediction) と呼び，その次元はターゲットの出力 (グランドトゥルース) と同じになります．

最初の層の畳み込み層が，1チャネルの入力，16チャネルの出力を持ち，カーネルサイズは3でストライドが1であることの注目してください．1チャネルの入力は，モデルに入力するグレイスケールの画像によるものです．カーネルサイズが 3x3 なのにはいくつか理由があります．まず，通常カーネルサイズは奇数に設定されます．これは，入力画像ピクセルは中央のピクセルの周辺に対称的に分布しているからです．1x1 カーネルでは周辺ぴく背うrの情報を何も取り込めなくなるので，カーネル操作としては小さすぎます．次に来るのは3ですが，5や7，27のようなカーネルは使わないのでしょうか？

28x28 の画像を 27x27 カーネルで畳み込めば，とても大雑把な特徴量が得られます．しかし，最も重要な視覚的特徴量は等しく局所的であり，したがって視覚的なパターンを抽出するのに，一度に周辺の数ピクセルだけを見るために小さいカーネルを使うというのは納得できます．CV 系の問題解決において，CNN で 3x3 のカーネルは最もよく使われるカーネルサイズの一つです．

ここでは，連続で 3x3 カーネルを用いて畳み込みを行っています．空間的なカバレッジの観点でいうと，これは1層の 5x5 カーネルを使った畳み込みと等価になります．しかし，ほとんどの場合で小さいカーネルを複数層重ねる方が好ましいとされています．なぜなら，ネットワークが深くなることでより少ないパラメータで複雑な特徴量を学習できるからです．

通常，出力側のチャネル数は入力側のチャネル数と同じか多くなります．初層の畳み込みでは1チャネルの入力を受け取り，16チャネルを出力しています．これは，基本的に入力画像から16種類の異なる情報を検出しようとしていることを意味しています．これら各チャネルは特徴量マップ (feature map) と呼ばれ，それぞれに特徴抽出を行うカーネルを持っています．2層目の畳み込みでチャネルを16から32に増やすことで，さらに多くの種類の特徴量を抽出しようとしていることを表しています．このようなチャネル数 (あるいは画像の深さ) を増やす方法は，CNN でよく用いられます．

最後に，カーネルサイズが3であることを考えれば，ストライドが1であることは合理的であると言えます．ストライドの値を大きくすると多くのピクセルをスキップすることになりますが，そんなことはやる意味がありません．しかし，もしカーネルサイズが100だった場合には，ストライドが10でも良いでしょう．ストライドを大きくすると畳み込み演算の回数自体は減りますが，カーネルが見られる領域は狭くなります．

````python
# 誤差逆伝播による学習を行う関数を定義
def train(model, device, train_dataloader, optim, epoch):
    model.train()
    for b_i, (X, y) in enumerate(train_dataloader):
        X, y = X.to(device), y.to(device)
        optim.zero_grad()
        pred_prob = model(X)
        loss = F.nll_loss(pred_prob, y) # nll = negative likelihood loss
        loss.backward()
        optim.step()
        if b_i % 10 == 0:
              print('epoch: {} [{}/{} ({:.0f}%)]\t trainingloss: {:.6f}'.format(
                  epoch,
                  b_i * len(X), 
                  len(train_dataloader.dataset),
                  100. * b_i / len(train_dataloader), 
                  loss.item()
              ))
````

このデータセットをバッチごとに繰り返し，指定されたデバイス上にデータセットのコピーを作成し，取得したデータをニューラルネットワークモデルで順伝播し，モデルの予測と正解データの間で損失関数を計算し，モデルの重みのチューニングに指定された最適化手法を適用し，そして10バッチごとに学習のログを表示します．この手順を1回行うと1エポック，つまりデータセット全体が1回読み込まれたことになります．

````python
# モデルの性能評価を行う関数を定義
def test(model, device, test_dataloader):
    model.eval()
    loss = 0
    success = 0
    with torch.no_grad():
        for X, y in test_dataloader:
            X, y = X.to(device), y.to(device)
            pred_prob = model(X)
            loss += F.nll_loss(pred_prob, y, reduction="sum")  # バッチごとにロスを合計
            pred = pred_prob.argmax(dim=1, keepdim=True)  # 予測されたラベルを取得
            success += pred.eq(y.view_as(pred)).sum().item()

    loss /= len(test_dataloader.dataset)
    print('\nTest dataset: Overall Loss: {:.4f}, OverallAccuracy: {}/{} ({:.0f}%)\n'.format(
        loss,
        success,
        len(test_dataloader.dataset),
        100. * success / len(test_dataloader.dataset)
    ))
````

この関数はほとんどが ```train``` 関数と同じ内容になっています．1点異なるのは，計算した損失関数から最適化手法によってモデルの重みをチューニングしていない点です．その代わり，テストバッチ全体についてテストエラーを計算するのに使っています．

````python
# PyTorch の DataLoader モジュールを使ってデータセットを読み込む
# 平均と標準偏差の値は学習用のデータセットですべての画像の画素から計算している

train_dataloader = torch.utils.data.DataLoader(
    datasets.MNIST("./data", train=True, download=True, 
                    transform=transforms.Compose([
                        transforms.ToTensor(),
                        transforms.Normalize((0.1302,), (0.3069,))
                    ])), 
    batch_size=32, shuffle=True
)

test_dataloader = torch.utils.data.DataLoader(
    datasets.MNIST('./data', train=False, download=True,
                    transform=transforms.Compose([
                        transforms.ToTensor(),
                        transforms.Normalize((0.1302,), (0.3069,))
                    ])),
    batch_size=500, shuffle=False
)
````

ここでは ```batch_size``` を32としていますが，よく使われる値です．しばしば，バッチサイズの選択にはトレードオフが存在します．バッチサイズをとても小さな値に設定すると，勾配の計算が頻繁に行われるために学習が遅くなり，非常にノイジーな勾配になってしまいます．一方，バッチサイズを大きく設定すると，勾配の計算を待つ必要があるため学習が遅くなります．1回の勾配更新の前に長く待つ価値はありません．より正確ではない勾配で頻繁に更新を行った方が，より良いパラメータに近づいていきます．

学習とテストデータセットの両方について，データセットを保存するローカルストレージの場所と，学習とテストの1回の実行を構成するデータインスタンスの数を決めるバッチサイズを指定します．また，バッチ間のデータサンプルの分布が均一になるように，学習データのインスタンスをランダムにシャッフルすることも指定します．最後に，データセットを平均値と標準偏差が指定された正規分布に正規化します．

````python
# 最適化手法の選定とデバイスの指定
torch.manual_seed(0)
device = torch.device("cpu")

model = ConvNet()
optimizer = optim.Adadelta(model.parameters(), lr=0.5)
````

ここでは，実行デバイスを CPU に設定しています．また，再現性を担保するために乱数のシードを指定しています．最適化スケジュールには AdaDelta を使い，学習率は 0.5 に設定しています．AdaDelta はスパースなデータを扱う場合に適した最適化スケジュールです．MNIST のすべてのピクセルが情報を持っているわけではないので，このデータセットはスパースなデータに該当します．気になる人は，Adam などほかのスケジューラで学習した時に，学習処理やモデルの性能にどのような影響が出るかを調べてみてください．

````python
# 指定したエポック数の回数だけ学習を実行し，各エポックごとに評価を行う
for epoch in range(1, 3):
    train(model, device, train_dataloader, optimizer, epoch)
    test(model, device, test_dataloader)
````

````python
# 目で見て正しく推論できていることを確認してみる
test_samples = enumerate(test_dataloader)
b_i, (sample_data, sample_targets) = next(test_samples)

plt.imshow(sample_data[0][0], cmap="gray", interpolation="none")
````

````python
print(f"Model prediction is : {model(sample_data).data.max(1)[1][0]}")
print(f"Ground truth is : {sample_targets[0]}")
````

予測では，```axis=1``` とした ```max``` 関数を使って確率が最大のクラスを最初に計算します．```max``` 関数は2つのリストを出力します．1つ目が ```sample_data``` に含まれるすべてのサンプルのクラス確率，2つ目が各サンプルのクラスラベルです．そのため，最初に2つ目のリストを選択し，続いてクラスラベルを選択しています．
先ほどの結果を見る限り，正しく予測できてそうです．```model()``` によって順伝播を行い，確率を出力しています．したがって，```max``` 関数で確率が最大のクラスを出力します．
